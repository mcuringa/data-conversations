```python
from fuzzywuzzy import fuzz

qry = "IS 88"

t = df.copy(deep=False)
# t["match"] = t.clean_name.apply(lambda sn: fuzz.partial_ratio(qry, sn))

t["match"] = t.short_name.apply(lambda sn: fuzz.ratio(qry, sn))
t = t.query("match > 50")

# t["match"] = t.clean_name.apply(lambda sn: fuzz.ratio(qry, sn))
t = t.sort_values(by=["match"], ascending=False)

t = t[["dbn","short_name", "clean_name", "school_name", "match"]]
t
# t.clean_name.unique()
```
