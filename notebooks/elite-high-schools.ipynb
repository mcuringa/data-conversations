```python
# 2019 high school directory
url = "https://data.cityofnewyork.us/resource/uq7m-95z8.csv?$limit=10000000"
df = pd.read_csv(url)
list(df.columns)
```

```python
df = df[
    ['dbn', 'school_name', 'year', 'total_enrollment',
           'female', 'female_1', 'male',
           'male_1', 'asian', 'asian_1', 'black', 'black_1', 'hispanic',
           'hispanic_1', 'multiple_race_categories', 'multiple_race_categories_1',
           'white', 'white_1', 'students_with_disabilities',
           'students_with_disabilities_1', 'english_language_learners',
           'english_language_learners_1', 'poverty', 'poverty_1',
           'hs', 'district',
           'school_num', 'boro', 'boro_name', 'school_type',
           'non_white', 'non_white_1', 'black_hispanic',
           'black_hispanic_1', 'white_asian', 'white_asian_1', 'non_white_asian',
           'non_white_asian_1'
    ]
]

# get high schools for the 2019-2020 school year
hs = df.query("hs==True and year == 2019")

# fix raw poverty counts which are strings to get an estimate
def pov(row):
        if "Below" in row.poverty:
            row.poverty = int(row.total_enrollment * .04)
        if "Above" in row.poverty:
            row.poverty = int(row.total_enrollment * .96)
        row.poverty = int(row.poverty)
        return row

hs = hs.apply(pov, axis=1)

hs.columns

```

```python
# the school IDs of the 9 specialized high schools (including laguardia)
spec_dbn = ["10X445", "02M475","14K449" , "13K430", "03M485", "05M692", "10X696", "28Q687", "31R605"]

# 9 top ranked high schools by US News, excluding specialized
us_news = ["25Q525", "02M545", "02M416", "30Q580", "02M418", "05M362", "01M539", "27Q323", "15K684"]


hs["specialized"] = hs.dbn.isin(spec_dbn)
hs["us_news"] = hs.dbn.isin(us_news)

hs.query("us_news==True")
```

```python
agg_funcs = {"dbn":"count", 
             "total_enrollment":"sum", 
             "black":"sum",
             "black_1":"mean",
             "white":"sum", 
             "white_1":"mean", 
             "hispanic":"sum",
             "hispanic_1":"mean",
             "asian":"sum",
             "hispanic_1":"mean",
             "students_with_disabilities":"sum",
             "hispanic_1":"mean",
             "poverty":"sum",
             "poverty_1":"mean",
             "english_language_learners":"sum",
             "hispanic_1":"mean",
             
            }

high_pov = hs.query("poverty_1 > .89").agg(agg_funcs)
high_pov["name"] = "high poverty"


wealthy = hs.query("poverty_1 < .40")
print(wealthy.sort_values(by=["poverty_1"])[["school_name", "white_1", "black_1", "hispanic_1", "asian_1"]])

wealthy = wealthy.agg(agg_funcs)
wealthy["name"] = "low poverty"


all_schools = hs.agg(agg_funcs)
all_schools["name"] = "all schools"
# agg.append(high_pov, ignore_index=False)

pd.DataFrame([all_schools, high_pov, wealthy])
```
