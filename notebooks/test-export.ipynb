```python
import pandas as pd
import schools
# df = schools.load_demographics()
# df.head()

df = pd.read_feather("combined-test-data.feather")
df.head()
```

```python
year = df.year.max()
print(year)
cols = list(df.columns)
cols = [c for c in cols if "_level_" not in c]
cols = [c for c in cols if c not in ['clean_name', 'short_name']]
def f(col):
    if "grade" not in col:
        return True
    if "grade_4" not in col and "grade_8" not in col:
        return False
    return True

cols = [c for c in cols if f(c)]
d4 = df.query("grade_4 > 0 and year == 2018")
d8 = df.query("grade_8 > 0 and year == 2018")
cols4 = [c for c in cols if "grade_8" not in c]
cols8 = [c for c in cols if "grade_4" not in c]
d4 = d4[cols4]
d8 = d8[cols8]

d4.to_csv("grade4-tests.csv")
d4.to_csv("grade8-tests.csv")
cols4
    
```
