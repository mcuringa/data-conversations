```python
def pov(row):
    try:
        float(row.poverty_1)
        return float(row.poverty_1)
    except:
        pass
    
    if row.poverty == "Below 5%":
        return .04
    if row.poverty == "Above 95%":
        return .96

    pct = str(row.poverty_1)[:-1]
    return float(pct) / 100


df.poverty_1 = df.apply(pov, axis=1)

t = df.query("year == 2019 and district==24")
t = t.sort_values(by=["poverty_1", "district"])[schools.SCHOOL_BASIC + schools.DEMO_GROUPS]
t.iloc[0].T
```
